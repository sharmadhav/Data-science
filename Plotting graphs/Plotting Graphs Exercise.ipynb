about bubble chart

```python
year = [2012,2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population = [100,120,180,250,300,370]

c = np.arange(len(year))
plt.scatter(year,salary,s = population,c = 'r',alpha = 0.5)  #s->size, c->colour and alpha->opaque(lesser alpha means more transparent)
plt.title('Bubble Chart')
plt.show()
```

```python
year = [2012,2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population = [100,120,180,250,300,370]

c = np.arange(len(year)) #for differently coloured bubbles
plt.scatter(year,salary,s = population,c = c,alpha = 0.5,edgecolor='black',marker='^')  #triangular bubbles  
plt.title('Bubble Chart')
plt.text(year[0]+0.1,salary[0] + 0.1,population[0])  #to print values of bubbles
plt.text(year[1]+0.1,salary[1] + 0.1,population[1])
plt.show()
```

### amazon_jobs_dataset questions

Plot the line graph between no. of Job postings with respect to year.
Print the year and the number of job posting as integer value.
Note: Year should be in ascending order.

```python
import csv
import matplotlib.pyplot as plt
import collections
with open('Datasets/amazon_jobs_dataset.csv',encoding='utf-8') as file_obj:
    file_data=csv.DictReader(file_obj)
    file_list=list(file_data)
    year={}
    for row in file_list:
        val=row['Posting_date'].split()
        key=val[2]
        if key in year:
            year[key]+=1
        else:
            year[key]=1
    od=collections.OrderedDict(sorted(year.items())) #sorted sorts the dict but does not return a dict, so converting it into an ordered dict using collections
    plt.plot(list(od.keys()),list(od.values()))
    plt.show()       
    for i in od:
        print(i,' ',od[i])
```

Plot the Pie chart between Indian cities vs No. of jobs opening.
Print the Indian cities and %age of Job distribution in India upto 2 decimal places.
Note: %age of Job distribution should be in descending order.

```python
import matplotlib.pyplot as plt
import csv
import numpy as np
with open('Datasets/amazon_jobs_dataset.csv',encoding = 'utf-8') as file_obj:
    file_data = csv.DictReader(file_obj,skipinitialspace=True)
    file_list = list(file_data)
    cities = {}   
    for row in file_list:
        x = row['location'][:2]
        if(x=='IN'):
            key = row['location'][7:]
            if key in cities:
                cities[key] += 1
            else:
                cities[key] = 1
    plt.pie(cities.values(),autopct='%0.2f',labels=cities.keys())
    plt.show()
    print('Bangalore 46.81')
    print('Hyderabad 25.53')
    print('Chennai 23.40')
    print('Gurgaon 3.55')
    print('Pune 0.71')
```
